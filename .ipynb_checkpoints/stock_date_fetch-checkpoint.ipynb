```python
import baostock as bs

lg = bs.login()
fields = "date, open, high, low, close, volume"
df_bs = bs.query_history_k_data("sh.000001", fields, start_date='2009-01-01', end_date='2019-01-10', frequency="d", adjustflag="2")
print(type(df_bs))
```

```python
data_list = []
while(df_bs.error_code == '0') & df_bs.next():
    data_list.append(df_bs.get_row_data())
```

```python
import pandas as pd

result = pd.DataFrame(data_list, columns=df_bs.fields)
result.close = result.close.astype('float64')
result.open = result.open.astype('float64')
result.low = result.low.astype('float64')
result.high = result.high.astype('float64')
result.volume = result.volume.astype('int64')
result.index = pd.to_datetime(result.date)
```

```python
#Baostock使用样例
import baostock as bs
import pandas as pd

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 6)

lg = bs.login()

rs = bs.query_history_k_data_plus("sh.600000",
    "date,code,open,high,low,close,preclose,volume,amount,adjustflag,turn,tradestatus,pctChg,isST",
    start_date='2017-07-01', end_date='2017-07-10',
    frequency="d", adjustflag="3")

data_list = []
while (rs.error_code == '0') & rs.next():
    # 获取一条记录，将记录合并在一起
    data_list.append(rs.get_row_data())
result = pd.DataFrame(data_list, columns=rs.fields)
result.index = pd.to_datetime(result.date)
result = result.drop(['date'], axis=1)
print(result)
```
